```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (assuming it has been created and saved as 'automobile_sales_data.csv')
df = pd.read_csv('automobile_sales_data.csv')

# Ensure the Date column is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract year from the Date
df['Year'] = df['Date'].dt.year

# Group by Year and Vehicle_Type, summing the Automobile_Sales
sales_by_vehicle_type = df.groupby(['Year', 'Vehicle_Type'])['Automobile_Sales'].sum().unstack()

# Plotting the line chart
plt.figure(figsize=(14, 7))
for vehicle_type in sales_by_vehicle_type.columns:
    plt.plot(sales_by_vehicle_type.index, sales_by_vehicle_type[vehicle_type], marker='o', label=vehicle_type)

plt.title('Automobile Sales Trends by Vehicle Type', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Total Automobile Sales', fontsize=14)
plt.xticks(sales_by_vehicle_type.index, rotation=45)
plt.grid()
plt.legend(title='Vehicle Type', fontsize=12)
plt.tight_layout()

# Save the plot as an image
plt.savefig('Line_Plot_Vehicle_Types.png')

# Show the plot
plt.show()
```
