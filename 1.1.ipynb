```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (assuming it has been created and saved as 'automobile_sales_data.csv')
df = pd.read_csv('automobile_sales_data.csv')

# Ensure the Date column is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract year from the Date
df['Year'] = df['Date'].dt.year

# Group by year and sum the Automobile_Sales
sales_per_year = df.groupby('Year')['Automobile_Sales'].sum().reset_index()

# Plotting the line chart
plt.figure(figsize=(12, 6))
plt.plot(sales_per_year['Year'], sales_per_year['Automobile_Sales'], marker='o')
plt.title('Automobile Sales Fluctuation from Year to Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Total Automobile Sales', fontsize=14)
plt.xticks(sales_per_year['Year'], rotation=45)
plt.grid()
plt.tight_layout()

# Save the plot as an image
plt.savefig('Line_Plot_1.png')

# Show the plot
plt.show()
```
