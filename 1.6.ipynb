```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (assuming it has been created and saved as 'automobile_sales_data.csv')
df = pd.read_csv('automobile_sales_data.csv')

# Filter data for recession periods
recession_data = df[df['Recession'] == 1]

# Create a scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(recession_data['Price'], recession_data['Automobile_Sales'], 
            alpha=0.6, edgecolors='w', linewidth=2)

# Adding titles and labels
plt.title('Correlation between Average Vehicle Price and Sales Volume During Recessions', fontsize=16)
plt.xlabel('Average Vehicle Price (USD)', fontsize=14)
plt.ylabel('Automobile Sales Volume', fontsize=14)
plt.grid(True)

# Save the plot as an image
plt.savefig('Scatter_Plot_Avg_Price_vs_Sales.png')

# Show the plot
plt.show()
```
